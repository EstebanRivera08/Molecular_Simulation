## Classes can also be run from a notebook 
Print output (and error messages) from the simulation can be seen from the "log console". Open "View-> Show Log Console", and set the console to "info". Note the first line `%matplotlib widget`, is needed by jupyter-lab in order to display the animation

```python
%matplotlib widget

from simul import Simul
from animatesimul import AnimateSimul


simulation = Simul(sample_time=0.01, sigma=0.15)
animate = AnimateSimul(simulation)
animate.go(nframes=100)
```
